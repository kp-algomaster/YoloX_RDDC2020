YOLO X implementation on Road Damage detection Dataset

The RDD2020 dataset contains 26336 road images collected from India, Japan, and the Czech Republic with more than 31000 instances of road damage. The dataset contains annotation for four damage categories: Longitudinal Cracks(D00), Transverse Cracks(D10), Alligator Cracks(D20), and Potholes(D40); and is intended for developing deep learning-based methods to detect and classify road damage automatically.

```python
import os
import random

trainval_percent = 0.1
train_percent = 0.9
xmlfilepath = '/content/gdrive/MyDrive/DL_Experiments/Road_Damage_Detection/Models/YOLOX-main/tools/datasets/VOCdevkit/VOC20/Annotations'
txtsavepath = '/content/gdrive/MyDrive/DL_Experiments/Road_Damage_Detection/Models/YOLOX-main/tools/datasets/VOCdevkit/VOC20/ImageSets'
total_xml = os.listdir(xmlfilepath)

num = len(total_xml)
list = range(num)
tv = int(num * trainval_percent)
tr = int(tv * train_percent)
trainval = random.sample(list, tv)
train = random.sample(trainval, tr)

ftest = open('/content/gdrive/MyDrive/DL_Experiments/Road_Damage_Detection/Models/YOLOX-main/tools/datasets/VOCdevkit/VOC20/ImageSets/Main/test.txt', 'w')
ftrain = open('/content/gdrive/MyDrive/DL_Experiments/Road_Damage_Detection/Models/YOLOX-main/tools/datasets/VOCdevkit/VOC20/ImageSets/Main/train.txt', 'w')

for i in list:
    name = total_xml[i][:-4] + '\n'
    if i in trainval:
        ftest.write(name)
    else:
        ftrain.write(name)

ftrain.close()
ftest.close()
```
